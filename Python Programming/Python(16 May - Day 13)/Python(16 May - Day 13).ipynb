# Problem 1:
### Problem Statement: Define a function to read data from a text file

```python
def readFileData(filename):
    f = open(filename, 'r')   # file is opened in read 'r' mode
    
    filedata = f.read() # reads all the test in the file
    f.close()   # file is closed by the file handler   
    return filedata   # filedata is being returned in form of a variable

readFileData('DataFiles/data.txt')
```

```python
def readFileData(filename):
    f = open(filename, 'r')   # file is opened in read 'r' mode
    
    filedata = f.readline()  # reads 1st line
    filedata += f.readline() # appends 2nd line to the 1st line
    f.close()
    
    return filedata   # filedata is being returned in form of a variable

readFileData('DataFiles/data.txt')
```

# Problem 3:
### Problem Statement: Define a function to generate a Marks data file for 1300 students such that each mark is entered in a new line. Marks range from 0 to 100 (inclusive) as random numbers

```python
from random import randint

def MarksDataFile(filename, students):    
    with open(filename, 'w') as f: 
        for i in range(0, students):
            marks = randint(0,100)
            f.write(str(marks) + '\n')
    return

MarksDataFile('DataFiles/MarksData.txt', 1300)
```

# Problem 4:
### Problem Statement: Generate a report on the marks data with the following indicators:
* highest mark:
* lowest mark:
* average mark:
* no of students with distinction(>80):
* no of students with first class(>60):
* no of students with second class(>50):
* no of students with third class(>40):
* no of students failed(<40):

```python
import timeit

def minMark(marks):
    min_m = marks[1]
    for i in range(0, len(marks)):
        if marks[i] < min_m:
            min_m = marks[i]
    return min_m

def maxmark(marks):
    max_m = marks[1]
    for i in range(0, len(marks)):
        if marks[i] > max_m:
            max_m = marks[i]
    return max_m

def avgMark(min_m, max_m):
    avg_m = (min_m + max_m)/2
    return avg_m

def generateMarksReport(MarksReport, MarksData):
    start = timeit.default_timer()
    marks, dist, first, second, third, fail = [], [], [], [], [], []
    
    with open(MarksData, 'r') as f:
        for line in f:
            marks.append(int(line))
            
    min_m = minMark(marks) 
    max_m = maxmark(marks)
    avg_m = avgMark(min_m, max_m)
    
    for i in range(0, len(marks)):
        if marks[i] > 80:
            dist.append(marks[i])
    for i in range(0, len(marks)):
        if marks[i] > 60 and marks[i] <= 80 :
            first.append(marks[i])
    for i in range(0, len(marks)):
        if marks[i] > 50 and marks[i] <= 60:
            second.append(marks[i])
    for i in range(0, len(marks)):
        if marks[i] > 40 and marks[i] <= 50:
            third.append(marks[i])
    for i in range(0, len(marks)):
        if marks[i] <= 40:
            fail.append(marks[i])
    
    with open(MarksReport, 'w') as f:
        f.write('Highest mark : ' + str(max_m) + '\n')
        f.write('Lowest mark : ' + str(min_m) + '\n')
        f.write('Average mark : ' + str(avg_m) + '\n')
        f.write('Number of students with distinction : '+ str(len(dist)) + '\n')
        f.write('Number of students with first class : '+ str(len(first)) + '\n')
        f.write('Number of students with second class : '+ str(len(second)) + '\n')
        f.write('Number of students with third class : '+ str(len(third)) + '\n')
        f.write('Number of students who failed : '+ str(len(fail)) + '\n')

    return marks
            
generateMarksReport('DataFiles/MarksReport.txt', 'DataFiles/MarksData.txt')
```
